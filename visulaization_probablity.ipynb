```python
import matplotlib.pyplot as plt

students = ["A", "B", "C", "D", "E"]
scores = [75, 85, 60, 90, 70]

plt.figure()
plt.bar(students, scores)
plt.title("Student Scores - Bar Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()

plt.figure()
plt.plot(students, scores, marker='o')
plt.title("Student Scores - Line Chart")
plt.xlabel("Students")
plt.ylabel("Scores")
plt.show()

passed = 30
total = 100
probability = passed / total

print("Probability of passing:", probability)
```
