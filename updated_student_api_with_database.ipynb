```python
from sqlalchemy import create_engine, Column, Integer, String
from sqlalchemy.orm import declarative_base, sessionmaker

DATABASE_URL = "sqlite:///./students.db"

engine = create_engine(DATABASE_URL)
SessionLocal = sessionmaker(bind=engine)
Base = declarative_base()
```

```python
class Student(Base):
    __tablename__ = "students"

    id = Column(Integer, primary_key=True, index=True)
    name = Column(String)
    age = Column(Integer)
    course = Column(String)

Base.metadata.create_all(bind=engine)
```

```python
db = SessionLocal()
```

```python
s1 = Student(name="Shivank", age=20, course="AI")
s2 = Student(name="Rahul", age=21, course="Data Science")

db.add_all([s1, s2])
db.commit()

print("Students added")
```

```python
students = db.query(Student).all()

for s in students:
    print(s.id, s.name, s.age, s.course)
```

```python
keyword = "shiv"

results = db.query(Student).filter(Student.name.contains(keyword)).all()

for r in results:
    print(r.id, r.name)
```

```python
student = db.query(Student).filter(Student.id == 1).first()

if student:
    student.course = "Machine Learning"
    db.commit()
    print("Updated")
```

```python
student = db.query(Student).filter(Student.id == 2).first()

if student:
    db.delete(student)
    db.commit()
    print("Deleted")
```

```python
db.close()
```
